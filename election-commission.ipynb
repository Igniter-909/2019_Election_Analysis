### Question 1: Which States/UTs and Constituencies had the highest and the lowest Voter Turnout?

```python
constituency_voter_turnout=mer.groupby("CONSTITUENCY").first()
tt=mer.groupby("CONSTITUENCY")[["TOTAL\nVOTES"]].sum()
constituency_voter_turnout=pd.merge(constituency_voter_turnout,tt,on="CONSTITUENCY",how="left")
constituency_voter_turnout["VOTER PERCENTAGE"]=(constituency_voter_turnout["TOTAL\nVOTES_y"]/constituency_voter_turnout["TOTAL ELECTORS"])*100
constituency_voter_turnout=constituency_voter_turnout[["STATE","TOTAL\nVOTES_y","TOTAL ELECTORS","VOTER PERCENTAGE"]].sort_values(by="VOTER PERCENTAGE",ascending=False)
constituency_voter_turnout
```

The constituency "DHUBRI" in ASSAM has the highest voter turnout of 89.67%. While, "ANANTNAG" in JAMMU AND KASHMIR has lowest voter turnout of  8.40%.

```python
state_voter_turnout=mer.groupby("STATE")[["TOTAL\nVOTES"]].sum()
rr=constituency_voter_turnout.groupby("STATE")[["TOTAL ELECTORS"]].sum()
tt=pd.merge(state_voter_turnout,rr,on="STATE",how="left")
tt["VOTER PERCENTAGE"]=((tt["TOTAL\nVOTES"]/tt["TOTAL ELECTORS"])*100)
tt=tt.sort_values(by="VOTER PERCENTAGE",ascending=False)
tt
```

Among the state and union territories, LAKSHADWEEP has highest voter turnout of 83.75%, while J&K has least voter turnout of 43.32%.

### Qestion 2: How many elected MPs have a criminal record and what is their party-wise distribution?

```python
elected=mer[mer["WINNER"]==1]
criminal_winning_mps = elected[elected['CRIMINAL\nCASES'] > 0]
No_of_MPs=len(criminal_winning_mps)
No_of_MPs
```

```python
dist=criminal_winning_mps.groupby("PARTY")[["CRIMINAL\nCASES"]].sum().sort_values(by="CRIMINAL\nCASES",ascending=False)
dist
```

```python
dist.sum()
```

In a total, 1029 criminal cases are registered against the elected MP's. The maximum cases are registered against members of BJP{403}, followed by INC{355}, SHS{35}.

### Question 3: What is the distribution of assets, liabilities, and net worth of the elected MPs?

```python
mer.columns
```

```python
elected[["0","1"]]=elected["ASSETS"].str.split("\n",expand=True)
elected[["2","3"]]=elected["LIABILITIES"].str.split("\n",expand=True)
elected.drop(columns=['1','3'], inplace=True)
distr=elected[["NAME","0","2"]]
distr.rename(columns={"0":"ASSETS","2":"LIABILITIES"},inplace=True)
distr["ASSETS"]=distr["ASSETS"].str.replace("Rs ","")
distr["LIABILITIES"]=distr["LIABILITIES"].str.replace("Rs ","")
distr
```

```python
distr["ASSETS"]=distr["ASSETS"].str.replace(",","").astype(int)
distr["LIABILITIES"]=distr["LIABILITIES"].str.replace(",","").astype(int)
```

```python
distr.info()
```

```python
distr["NET WORTH"]=distr["ASSETS"]-distr["LIABILITIES"]
```

```python
distr.sort_values(by="NET WORTH",ascending=False)
```

**Among the elected MP's, "NAKUL KAMAL NATH" has the highest net worth of 659+ crore.**

### Question 4: Which 10 constituencies saw the Most One-sided and the Closest victories respectively?

```python
elect=mer[mer["WINNER"]==1]
```

```python
one_sided=elect.sort_values(by="OVER TOTAL ELECTORS \nIN CONSTITUENCY",ascending=False)
one_sided=one_sided[["NAME","CONSTITUENCY","OVER TOTAL ELECTORS \nIN CONSTITUENCY"]].head(10)
one_sided
```

```python
non_elect=mer[mer["WINNER"]==0].groupby("CONSTITUENCY")[["OVER TOTAL ELECTORS \nIN CONSTITUENCY"]].max()
win=mer[mer["WINNER"]==1].groupby("CONSTITUENCY")[["OVER TOTAL ELECTORS \nIN CONSTITUENCY"]].max()
```

```python
closet=pd.merge(win,non_elect,on="CONSTITUENCY",how="left")
closet["MARGIN"]=closet["OVER TOTAL ELECTORS \nIN CONSTITUENCY_x"]-closet["OVER TOTAL ELECTORS \nIN CONSTITUENCY_y"]
```

```python
closet.sort_values(by="MARGIN",ascending=True).head(10)
```
